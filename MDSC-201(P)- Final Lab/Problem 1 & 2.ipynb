```r
library(readxl)
data <- read_excel('/RM-Final Lab.xlsx')
data
```

```r
print(colnames(data))
```

```r
correlation_coefficient <- cor(data$Months, data$`#Songs`)
print(correlation_coefficient)
```

```r
# Performing Simple Linear Regression
model <- lm(`#Songs` ~ Months, data)

# Plotting the scatter plot of the data
plot(data$Months, data$`#Songs`, main="Scatter Plot with Fitted Line", xlab="Months", ylab="#Songs")

# Add the regression line to the scatter plot
abline(model, col="red")
legend("topright", legend=c("Data", "Fitted Line"), col=c("black", "red"), lty=1:1, cex=0.8)
```

**2nd Question**

```r
data1 <- read_excel('/second_question.xlsx')
model <- lm(GGPA ~ RE_Total + UGPA, data1)
summary(model)
```

```r
# Predict GGPA using the model
predicted <- predict(model)

# Plot actual versus predicted GGPA
plot(data1$GGPA, predicted, main = "Actual vs Predicted GGPA",
     xlab = "Actual GGPA", ylab = "Predicted GGPA", pch = 19)
abline(0, 1, col = "red")
```

Coefficients of the Regression Line

```r
coefficients <- coef(model)
print(coefficients)
```

P-Values

```r
summary(model)$coefficients[, "Pr(>|t|)"]
```
